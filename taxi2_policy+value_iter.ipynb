```python
# ************************
# Policy Iteration, small case, using Utility5
# ************************
env = gym.make('Taxi-v2')

result = defaultdict(list)

for gamma in [0.5 + 0.025*i for i in range(0, 20)]:
    t0 = time.time()
    pi, V = u5.policy_iteration(env, policy_eval_fn=u5.policy_eval, gamma=gamma)
    tt = time.time()-t0
    
    print("gamma = {:.2f}, time elapsed: {:.4f}".format(gamma, tt))
    result['V'].append(V)
    result['pi'].append(pi)
    result['optimal_steps'].append(np.mean([u5.count(pi, env) for i in range(100)]))
    result['time'].append(tt)
    result['gamma'].append(gamma)
    result['V0'].append(V[0])
    result['V100'].append(V[100])
    result['V200'].append(V[200])
    result['V300'].append(V[300])
    result['V400'].append(V[400])
    
result_policy_iter_small_u5_2 = result
np.save('result_policy_iter_small_u5_2', result_policy_iter_small_u5_2)

env.close()
```
